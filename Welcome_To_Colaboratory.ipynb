```
""" 

Matplotlib = todo el paquete de visualizacion de python
pyplot = modulo en el paquete de matplotlib. Proporciona interfaz que permite crear figuras y ehes de forma implicita y automatica.
pylab = modulo que se instala junto al paquete de matplotlib. Ya no se recomienda su uso.

1. Primero importamos

2. Preparar datos

3. Usar funciones

"""

import matplotlib.pyplot as plt
import numpy as np
```

Principales tramas:

- diagrama de linea
- grafico de barras
- histograma
- grafico de dispersion
- grafico circular 

```
# diagrama de linea

# Datos de ejemplo
x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
x2 = [2, 5, 8]
y2 = [3, 4, 3]

# Crear el diagrama de líneas
plt.plot(x1, y1, label='linea 1', linewidth = 4, color = 'blue')
plt.plot(x2, y2, label='linea 2', linewidth = 4, color = 'green')

# Etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Diagrama de líneas con Matplotlib')

# Mostrar el gráfico
plt.legend()
plt.show()
```

```
##graficos de barra

# Datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [3, 7, 2, 5, 8]

# Crear el gráfico de barras
plt.bar(categorias, valores)

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de barras con Matplotlib')

# Mostrar el gráfico
plt.show()
```

```
#histogramas

# Generar datos de ejemplo con una distribución normal
np.random.seed(42)
datos = np.random.normal(50, 10, size=500)

# Crear el histograma
plt.hist(datos, bins=20, edgecolor='pink')

# Etiquetas y título
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de una distribución normal con Matplotlib')

# Mostrar el gráfico
plt.show()

```

```
#grafica de dispersion 

# Generar datos aleatorios
np.random.seed(42)
x1 = np.random.rand(50)
y1 = np.random.rand(50)
x2 = np.random.rand(50)
y2 = np.random.rand(50)

# Crear el gráfico de dispersión con diferentes colores
plt.scatter(x1, y1, color='green', label='Grupo 1')
plt.scatter(x2, y2, color='black', label='Grupo 2')

# Etiquetas, título y leyenda
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de dispersión con datos aleatorios y diferentes colores')
plt.legend()

# Mostrar el gráfico
plt.show()

```

```
#grafico circular

# Datos de ejemplo
edades = ['A', 'B', 'C', 'D']
valores = [15, 30, 25, 30]

# Crear el gráfico circular
plt.pie(valores, labels=edades, autopct='%1.1f%%', startangle=90)

# Asegurar que el gráfico sea un círculo perfecto
plt.axis('equal')

# Título
plt.title('Gráfico circular con Matplotlib')

# Mostrar el gráfico
plt.show()
```
